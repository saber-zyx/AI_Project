### Xử lý bảng transaction

```python
transaction.isna().sum()
```

```python
# Xóa cột osVer khỏi DataFrame
transaction = transaction.drop(columns=['osVer'])

# Kiểm tra lại để đảm bảo cột đã bị xóa
print(transaction.columns)
```

```python
# Xóa cột deviceModel khỏi DataFrame
transaction = transaction.drop(columns=['deviceModel'])

# Kiểm tra lại để đảm bảo cột đã bị xóa
print(transaction.columns)
```

```python
transaction['appID'] = transaction['appID'].fillna('unknown')
```

```python
transaction['cashbackTime'] = transaction['cashbackTime'].fillna(transaction['reqDate'])
```

```python
print("Giá trị và tần suất trong cột platform:")
print(transaction['platform'].value_counts(dropna=False))
```

```python
# 1️⃣ Tạo bảng map giữa sender và platform trong bảng transfer
platform_map = (
    transfer
    .dropna(subset=['platform'])  # chỉ lấy dòng có platform
    .groupby('sender')['platform']
    .agg(lambda x: x.mode()[0] if not x.mode().empty else None)  # lấy platform phổ biến nhất của user
    .reset_index()
)

# 2️⃣ Đổi tên để join dễ hơn
platform_map.rename(columns={'sender': 'userID', 'platform': 'platform_from_transfer'}, inplace=True)

# 3️⃣ Join với transaction
transaction = transaction.merge(platform_map, on='userID', how='left')

# 4️⃣ Fill platform trong transaction bằng dữ liệu từ transfer (nếu thiếu)
transaction['platform'] = transaction['platform'].fillna(transaction['platform_from_transfer'])

# 5️⃣ Xóa cột phụ sau khi fill xong
transaction.drop(columns=['platform_from_transfer'], inplace=True)
```

```python
transaction['platform'] = transaction['platform'].fillna('unknown')
```

```python
print("Giá trị và tần suất trong cột platform:")
print(transaction['platform'].value_counts(dropna=False))
```

```python
# 1️⃣ Tạo mapping giữa sender và deviceID từ bảng transfer
device_map = (
    transfer
    .dropna(subset=['deviceID'])  # chỉ lấy dòng có deviceID
    .groupby('sender')['deviceID']
    .agg(lambda x: x.mode()[0] if not x.mode().empty else None)  # lấy deviceID phổ biến nhất nếu có nhiều
    .reset_index()
)

# 2️⃣ Đổi tên cột để join dễ hơn
device_map.rename(columns={'sender': 'userID', 'deviceID': 'deviceID_from_transfer'}, inplace=True)

# 3️⃣ Join với transaction
transaction = transaction.merge(device_map, on='userID', how='left')

# 4️⃣ Fill giá trị thiếu trong transaction.deviceID bằng deviceID lấy từ transfer
transaction['deviceID'] = transaction['deviceID'].fillna(transaction['deviceID_from_transfer'])

# 5️⃣ Xóa cột phụ sau khi fill xong
transaction.drop(columns=['deviceID_from_transfer'], inplace=True)

# 6️⃣ Kiểm tra lại kết quả
print("Số lượng giá trị thiếu deviceID sau khi fill:", transaction['deviceID'].isna().sum())
```
